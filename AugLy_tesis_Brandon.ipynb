```python
# Imports needed to start the augmentation
import augly.image as imaugs
import PIL
import os
```

```python
# Functiton to copy the txt files to the folder of each augmentation
from shutil import copy

def copy_text_files(folder_name, text_list):
  source_path_prefix = os.path.join("/content", "images_to_train")
  destination_path_prefix = os.path.join("/mydrive", "augmented_images")
  for text in text_list:
    source = os.path.join(source_path_prefix, text)
    destination = os.path.join(destination_path_prefix, folder_name, text[:-4]+'__'+folder_name+text[-4:])
    copy(source, destination)
```

```python
# Getting only the txt files
import glob

%cd /content/images_to_train/
text_list = glob.glob("*.txt")
text_list.sort()
images_list.sort()
print(len(text_list))
print(len(images_list))
for i in range(10):
  print(text_list[i])
```

## Single functions

### Blur function

```python
# Create the directory in Google Drive to store the output of the blur function
!mkdir "/mydrive/augmented_images/blured" 
```

```python
# Calling the blur function with a radius of 1.875
output_path_prefix = os.path.join("/mydrive", "augmented_images", "blured")
for image in images_list:
  image_path = os.path.join(image)
  output_path = os.path.join(output_path_prefix, image[:-4]+"__blured"+image[-4:])
  print(image_path)
  meta_blur = []
  aug_image = imaugs.blur(image_path, radius=1.875, output_path=output_path, metadata=meta_blur)
```

```python
copy_text_files("blured", text_list)
```
